```python
# Install required libraries
!pip install gradio requests beautifulsoup4 transformers gtts

import requests
import gradio as gr
from bs4 import BeautifulSoup
from transformers import pipeline
from gtts import gTTS
import os

# Function to extract news using BeautifulSoup
def extract_news(company):
    search_url = f'https://www.bing.com/news/search?q={company}'
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    articles = []
    for item in soup.select('div.news-card')[:5]:  # Adjust selector as needed
        title_element = item.find('a', {'class': 'title'})
        title_text = title_element.get_text() if title_element else "No title available"
        link = title_element['href'] if title_element else "No link"
        summary_element = item.find('div', {'class': 'snippet'})
        summary = summary_element.get_text() if summary_element else "No summary available"

        articles.append({
            'title': title_text,
            'link': link,
            'summary': summary
        })

    return articles

# Function to generate sentiment report and Hindi audio
def generate_report(company):
    articles = extract_news(company)
    sentiment_pipeline = pipeline("sentiment-analysis")

    report = ""
    audio_files = []
    for index, article in enumerate(articles):
        title = article['title']
        summary = article['summary']
        link = article['link']

        # Perform Sentiment Analysis
        sentiment = sentiment_pipeline(summary)[0]
        sentiment_label = sentiment['label']
        sentiment_score = sentiment['score']

        # Add to report
        report += f"### {title}\n"
        report += f"**Link:** [Read More]({link})\n"
        report += f"**Summary:** {summary}\n"
        report += f"**Sentiment:** {sentiment_label} (Score: {sentiment_score:.2f})\n\n"

        # Convert summary to Hindi speech
        tts = gTTS(text=summary, lang="hi")
        audio_path = f"news_summary_{index}.mp3"
        tts.save(audio_path)
        audio_files.append(audio_path)

    # Return report and audio files
    return report, audio_files

# Gradio Interface
def gradio_interface(company):
    report, audio_files = generate_report(company)
    audio_outputs = [gr.Audio(audio_file, format="mp3") for audio_file in audio_files]
    return report, *audio_outputs

# Define Gradio inputs and outputs
inputs = gr.Textbox(label="Enter Company Name")
outputs = [
    gr.Markdown(label="Sentiment Report"),
    gr.Audio(label="Audio 1", format="mp3"),
    gr.Audio(label="Audio 2", format="mp3"),
    gr.Audio(label="Audio 3", format="mp3"),
    gr.Audio(label="Audio 4", format="mp3"),
    gr.Audio(label="Audio 5", format="mp3")
]

# Launch Gradio app
iface = gr.Interface(
    fn=gradio_interface,
    inputs=inputs,
    outputs=outputs,
    title="ğŸ“° Company News Sentiment Analyzer",
    description="Enter a company name to fetch news articles, analyze sentiment, and listen to summaries in Hindi."
)

iface.launch(debug=True)
```
